```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
#Create Engine
rds_connection_string = "postgres:postgres@localhost:5432/ETL_Challenge"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
json_file = "salary.json"
df = pd.read_json(json_file)
df.head(5)
```

```python
salary_df = df.rename(columns={"id":"ID","player":"Player","season17_18":"Salary","tm":"Team"})
salary_df["Salary"] = salary_df["Salary"].map("${:,.2f}".format)
salary_df.head(5)
```

```python
csv_file = "Resources/2017_nba_season_stats.csv"
stats = pd.read_csv(csv_file)
stats
```

```python
stats.to_sql(name='stats', con=engine, if_exists='append', index=False)
```

```python
salary_df.to_sql(name='salary', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('select * from stats', con=engine).head()
```

```python
pd.read_sql_query('select * from salary', con=engine).head()
```

```python
combined_info = pd.merge(stats,salary_df, on='Player')
combined_info
```

```python
combined_info.to_sql(name='total_info', con=engine, if_exists='append', index=False)
```
